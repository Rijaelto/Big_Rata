# Serie temporal

```python
r = np.cumsum(np.random.randn(1000))
s = np.cumsum(np.random.randn(1000))
```

## Estilo matlab

```python
plt.plot(r, label= "r")
plt.plot(s, label = "s")
plt.xlabel("año", fontsize = 20, rotation = 45)
plt.ylabel("Temperatura media")
plt.title("Europa")
plt.suptitle("Temperatura")
plt.legend()
plt.grid()
plt.show()
```

## Orientado a objetos

```python
fig,ax = plt.subplots()
ax.plot(r, label= "r")
ax.plot(s, label = "s")
ax.set_xlabel("año", fontsize = 20, rotation = 45)
ax.set_ylabel("Temperatura media")
ax.set_title("Temperatura")
ax.legend()
ax.grid()
plt.show()
```

```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

base = datetime.datetime(2005, 2, 1)
dates = np.array([base + datetime.timedelta(hours=(2 * i))
                  for i in range(732)])
N = len(dates)
np.random.seed(19680801)
y = np.cumsum(np.random.randn(N))

fig, axs = plt.subplots(3, 1, constrained_layout=True, figsize=(6, 6))
lims = [(np.datetime64('2005-02'), np.datetime64('2005-04')),
        (np.datetime64('2005-02-03'), np.datetime64('2005-02-15')),
        (np.datetime64('2005-02-03 11:00'), np.datetime64('2005-02-04 13:20'))]
for nn, ax in enumerate(axs):
    ax.plot(dates, y)
    ax.set_xlim(lims[nn])
    # rotate_labels...
    for label in ax.get_xticklabels():
        label.set_rotation(40)
        label.set_horizontalalignment('right')
axs[0].set_title('Default Date Formatter')
plt.show()
```
