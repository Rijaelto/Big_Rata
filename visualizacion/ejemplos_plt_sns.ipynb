# Grafico rapido

```python
serie = [3,25,48,9,64,5]
```

## Plot

```python
plt.plot(serie)
plt.show()
```

## Pie

```python
plt.pie(serie)
plt.show()
```

## Scatter

```python
tips = sns.load_dataset("tips")
```

```python
plt.scatter(tips.total_bill,tips.tip);
```

```python
tips.day.unique()
```

```python
sun = tips.loc[tips.day == "Sun"]
plt.scatter(sun.total_bill,sun.tip, label ="sunday" , c= "red")
plt.legend()
plt.show()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, hue =tips.day);
```

## catplot

```python
sns.violinplot(x = "day", y = "tip", data = tips)
```

# Serie temporal

```python
r = np.cumsum(np.random.randn(1000))
s = np.cumsum(np.random.randn(1000))
```

## Estilo matlab

```python
plt.plot(r, label= "r")
plt.plot(s, label = "s")
plt.xlabel("ao", fontsize = 20, rotation = 45)
plt.ylabel("Temperatura media")
plt.title("Europa")
plt.suptitle("Temperatura")
plt.legend()
plt.grid()
plt.show()
```

## Orientado a objetos

```python
fig,ax = plt.subplots()
ax.plot(r, label= "r")
ax.plot(s, label = "s")
ax.set_xlabel("ao", fontsize = 20, rotation = 45)
ax.set_ylabel("Temperatura media")
ax.set_title("Temperatura")
ax.legend()
ax.grid()
plt.show()
```
