```python
from sklearn.decomposition import PCA
irisData = iris.iloc[:,0:4] 
pca = PCA(n_components=2)
pca.fit(irisData)
IrisPCA = pca.transform(irisData)
IrisOneParameter = pd.DataFrame(data=IrisPCA, columns = ["x","y"])
IrisOneParameter = IrisOneParameter.join(iris.iloc[:,4:5])
fig, ax = plt.subplots(figsize=(8,8))
sns.scatterplot(x = "x", y = "y", data = IrisOneParameter, hue="species",ax = ax);
```
