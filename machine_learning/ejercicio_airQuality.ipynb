SacarÃ© la media de una ventana centrada en cada uno de los valores, de manera que tratare de obtener valores para los campos vacios que sean la media de los valores de los dÃ­as anteriores y siguientes. 

```python
def outlayers_media(df,column):
    rollingdf = df.rolling(5,center=True,min_periods=1).mean()
    rollingdf = rollingdf.rename(columns={column: column+"_rolling"})
    auxdf = pd.concat([df, rollingdf],axis=1)
    auxdf['aux']=df[column].mask(pd.isnull, rollingdf[column+"_rolling"])
    final = auxdf[['aux']]
    final = final.rename(columns={'aux': column})
    return final
```

La ventana no es muy grande para no coger datos muy alejados de la medida de ese dÃ­a, realmente no se si esto tiene sentido para el dioxido de carbono, tendrÃ­a que estudiar realmente el comportamiento del dioxido de carbono en la atmosfera y en las ciudades. PodrÃ­a quizÃ¡s emplearse otro acercamiento que fuera mÃ¡s adecuado para esto, pero para una primera aproximaciÃ³n puede ser Ãºtil. 

```python
while no1[no1['NOx(GT)'].isnull()].shape[0] > 0:
    no1 = outlayers_media(no1,'NOx(GT)')
```

Hago lo mismo para la columna NO2(GT) y CO(GT)

```python
no2 = data_cleaning_3[["NO2(GT)"]]
while no2[no2['NO2(GT)'].isnull()].shape[0] > 0:
    no2 = outlayers_media(no2,'NO2(GT)')
```

```python
no3 = data_cleaning_3[["CO(GT)"]]
while no3[no3['CO(GT)'].isnull()].shape[0] > 0:
    no3 = outlayers_media(no3,'CO(GT)')
```
